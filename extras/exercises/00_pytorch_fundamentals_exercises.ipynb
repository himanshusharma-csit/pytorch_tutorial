### 2. Create a random tensor with shape `(7, 7)`.


````python
# Import torch
import torch

# Create random tensor
random_tensor = torch.rand(size=[7,7])
````

### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 7)` (hint: you may have to transpose the second tensor).

````python
# Create another random tensor
another_tensor = torch.rand(size=[1,7])

# Perform matrix multiplication
# Using @ and matmul
result_1 = random_tensor @ another_tensor.T
result_2 = random_tensor.matmul(another_tensor.T)
result_3 = torch.matmul(random_tensor, another_tensor.T)

print(result_1)
print(result_2)
print(result_3)
````


### 6. Create two random tensors of shape `(2, 3)` and send them both to the GPU (you'll need access to a GPU for this). Set `torch.manual_seed(1234)` when creating the tensors (this doesn't have to be the GPU random seed). The output should be something like:


### 7. Perform a matrix multiplication on the tensors you created in 6 (again, you may have to adjust the shapes of one of the tensors).

### 8. Find the maximum and minimum values of the output of 7.

````python
# Find max
max_1 = result_1.max()
max_2 = torch.max(result_1)
print("Max: ", max_1, max_2)

# Find min
min_1 = result_1.min()
min_2 = torch.min(result_1)
print("Min: ", min_1, min_2)
````

### 9. Find the maximum and minimum index values of the output of 7.

````python
# Find arg max
max_arg1 = result_1.argmax()
max_arg2 = torch.argmax(result_1)
print("Arg Max: ", max_arg1.item(), max_arg2.item())

# Find arg min
min_arg1 = result_1.argmin()
min_arg2 = torch.argmin(result_1)
print("Arg Min: ", min_arg1.item(), min_arg2.item())
````

